```r
# listing files
get_files = function(x, base_dir){
    files = list_files(file.path(base_dir, x), 'features.tsv.gz')
    data.frame(collection = x, file = gsub('/+', '/', files))
}

base_dir = file.path(work_dir, 'n9k-train_original')
feat_files = dirs %>%
    lapply(get_files, base_dir=base_dir) 
feat_files 
```

```r
# formatting
cols = c('X1', 'ebio', 'abt3_projects', 'databases_no-backup', 'DeepMAsED', 
         'version2', 'resmico-sm', 'ftp', 'X2', 'X3', 'error_profile', 
         'features', 'richness', 'abundance_distribution', 'rep', 'read_length', 'read_depth', 
         'assembler', 'basename')
feat_files[[1]] = feat_files[[1]] %>%
    separate.(file, cols, '/', remove=FALSE) %>%
    mutate.(insert_size = 'mean270-stdev50')
feat_files[[1]]
```

```r
# formatting
cols = c('X1', 'ebio', 'abt3_projects', 'databases_no-backup', 'DeepMAsED', 
         'version2', 'resmico-sm', 'ftp', 'X2', 'X3',
         'features', 'richness', 'abundance_distribution', 'rep', 'read_length', 'read_depth', 
         'assembler', 'basename')

# formatting
feat_files[[2]] = feat_files[[2]] %>%
    separate.(file, cols, '/', remove=FALSE) %>%
    mutate.(error_profile = 'normal_err',
            insert_size = 'mean350-stdev75')
feat_files[[3]] = feat_files[[3]] %>%
    separate.(file, cols, '/', remove=FALSE) %>%
    mutate.(error_profile = 'cami_err',
            insert_size = 'mean350-stdev75')
feat_files[[4]] = feat_files[[4]] %>%
    separate.(file, cols, '/', remove=FALSE) %>%
    mutate.(error_profile = 'cami_err',
            insert_size = 'mean350-stdev75') %>%
    mutate.(rep = as.Num(rep) + 1)

feat_files = feat_files %>%
    rbindlist(use.names=TRUE)
feat_files
```

```r
# creating standardized dest
feat_files = feat_files %>%
    rename.('src_file' = file) %>%
    mutate.(X2 = 'n9k-train') %>%
    mutate.(dest_file = paste(X1, ebio, abt3_projects, `databases_no-backup`, 
                              DeepMAsED, version2, `resmico-sm`, ftp, X2,
                              features, error_profile, richness, abundance_distribution, 
                              insert_size, rep, read_length, 
                              read_depth, assembler, basename, sep='/')) %>%
    select.(collection, src_file, dest_file, error_profile, richness, abundance_distribution, 
            insert_size, rep, read_length, read_depth, assembler, basename)
feat_files
```

### Checks

```r
# checks: same src-dest?
feat_files %>%
    mutate.(same_file = src_file == dest_file) %>%
    filter.(same_file == TRUE)
```

```r
# checks: duplicate dest?
feat_files %>%
    summarize.(n = n.(), .by = c(dest_file)) %>%
    filter.(n > 1)
```

# Copying files

```r
copy_file = function(x, src_dir, dest_dir){
    src = file.path(src_dir, x)
    dest = file.path(dest_dir, x)
    if(src == dest){
        stop('src == dest')
    }
    if(file.exists(dest)){
        file.remove(dest)
    }
    file.copy(src, dest)  
}
```

```r
copy_files = function(x){
    dest_dir = dirname(x['dest_file'])
    make_dir(dest_dir, quiet=TRUE)  
    src_dir = dirname(x['src_file'])
    c('features_binary', 'features.tsv.gz', 'stats', 'toc') %>%
        lapply(copy_file, src_dir=src_dir, dest_dir=dest_dir)
}

# copying files
res = feat_files %>%
    apply(1, copy_files)
res %>% length
```

```r
# writing feature table
F = file.path(work_dir, 'n9k-train', 'feature_files.tsv')
feat_files %>%
    rename.('feature_file' = dest_file) %>%
    select.(richness, abundance_distribution, insert_size, error_profile, 
            rep, read_length, read_depth, assembler, feature_file) %>%
    mutate.(feature_file = gsub('.+/resmico-sm/ftp/n9k-train/', '', feature_file)) %>%
    write_table(F)
```
